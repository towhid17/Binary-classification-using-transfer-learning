```python
from keras.models import load_model
import os
import cv2
from PIL import Image
import numpy as np

# load model
model = load_model('b64_oA_d50_ovs_150ep.h5')

```

```python
image_directory = 'archive/'
SIZE = 256
testimg = [] 
label = []

testing = os.listdir(image_directory + 'test/')
for i, image_name in enumerate(testing): 
    if (image_name.split('.')[1] == 'png'):
        image = cv2.imread(image_directory + 'test/' + image_name)
        image = Image.fromarray(image, 'RGB')
        image = image.resize((SIZE, SIZE))
        testimg.append(np.array(image))
        label.append(0)
```

```python
testimg = np.array(testimg)
print(testimg.shape)
img = testimg[109]
input_img = np.expand_dims(img, axis=0)
# predict
pred = model.predict(input_img)

#plot the image
import matplotlib.pyplot as plt
plt.imshow(img)

print(pred)
```
